##### **02 Fish Weight 데이터 (ols)**
물고기의 무게를 다중회귀분석으로 분석하고자 한다. 데이터의 구성은 아래 표와 같다. Weight를 Y변수로 Length, Height, Width, FinSize를 X변수로 하는 다중회귀분석을 수행하고 아래에 답하시오.    

|id |Length |Height |Width |FinSize |Weight |
|---|-------|-------|------|--------|-------|
|1  |63.91  |28.72  |3.46  |18.22   |194.32 |

(1) Length의 회귀계수를 구하고, 소수 둘째 자리로 반올림하시오.    
(2) Height의 회귀계수의 P—value를 구하시오.    
(3) Width의 회귀계수가 P-value 0.05를 기준으로 했을 때 통계적으로 의미가 있는지를 평가하시오. 의미가 있는 경우엔 Y, 없는 경우엔 'N'을 출력하시오.

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/eyejjang/BigData/main/dataset_yemoon/datasets/Part5/ex6/06.03.02-fish_weight_data.csv")
print(df)
```

```python
#Length의 회귀계수를 구하고, 소수 둘째 자리로 반올림하시오.
dt = df.copy()
dt.pop('id')
#print(dt.head())
import statsmodels.formula.api as smf
model = smf.ols("Weight ~ Length+Height+Width+FinSize", dt).fit()
print(round(model.params['Length'], 2))
print('*'*100)
print(model.summary())
```

```python
#Height의 회귀계수의 p—value를 구하시오.
print(model.pvalues['Height'])
```

```python
#Width의 회귀계수가 p-value 0.05를 기준으로 했을 때 통계적으로 의미가 있는지를 평가하시오. 의미가 있는 경우엔 'Y', 없는 경우엔 'N'을 출력하시오.
if model.pvalues['Height'] < 0.05 :
    result = 'Y'
else :
    result = 'N'
print(result)
```

##### **01 Gender Prediction 데이터 (logit)**
키(Height)，몸무게(Weight), 발 사이즈(Shoe Size)를 이용하여 남녀(Sex) 성별을 분류하는 로지스틱 회귀모형을 만들고자 한다. 주어진 데이터 세트를 이용하여 로지스틱 회귀모형을 만들고 아래의 질문에 답하시오. 단, 주어진 데이터 세트의 2/3를 훈련용으로, 1/3을 테스트용으로 분할하여 작업하시오.    
(1) 훈련용 데이터를 이용하여 ‘Height’, ‘Weight’, ‘Shoe Size’를 설명변수로 사용하는 로지스틱 회귀모형을 만들고 Weight의 오즈비를 계산하시오.    
(2) 모델의 로짓우도(log-likelihood)를 구하시오.    
(3) (2)에서 만든 모델로 평가용 데이터를 예측한 결과와 실제값의 오차율을 구하시오.    

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/eyejjang/BigData/main/dataset_yemoon/datasets/Part5/ex7/07.03.01-gender_prediction_dataset.csv')
print(df)
df.describe(include='all').T
```

```python
dt = df.copy()
dt['Gender'] = dt['Gender'].map({'Male':1, 'Female':0}) # np.where 사용 가능. 예제 찾아보기
print(dt.head())
print('*'*100)
print(dt['Gender'].unique())
```

```python
#데이터 분리 하기
from sklearn.model_selection import train_test_split
y = dt['Gender']
train, test = train_test_split(dt, test_size=1/3, shuffle=False)
print(train.shape, test.shape)
```

```python
#훈련용 데이터를 이용하여 'Height', 'Weight', 'ShoeSize'를 설명변수로 사용하는 로지스틱 회귀모형을 만들고 Weight의 오즈비를 계산하시오.
import statsmodels.formula.api as smf
lm = smf.logit("Gender~Height+Weight+ShoeSize", train).fit()
print(lm.summary())
```

```python
#Weight의 odds 비 구하기 (odds 비 시 #오즈는 성공할 확률과 실패할 확률의 비이다.
#오즈비는 X변수의 값이 1만큼 늘어나면 y값에 오즈비(ex. 1.5배) 만큼 영향을 준다는 것이다.
#np.exp는 밑수인 a가 자연 상수 'e'(약 2.71828》이고. 이에 대해 입력값 n제곱을 한 것을 의미한다.
#지수를 얻기 위한 것으로 회귀계수(coef)를 지수로 변환하면 오즈비가 된다.
import numpy as np
odds_rto = np.exp(lm.params)
print(odds_rto['Weight'])
```

```python
#모델의 로그우도(log-likelihood)를 구하시오.
print(lm.llf)
```

```python
#평가용 데이터로 예측하기
y_pred = (lm.predict(test[['Height','Weight','ShoeSize']]) >= 0.5).astype(int)
```

```python
#(2)에서 만든 모델로 평가용 데이터를 예측한 결과와 실제값의 오차율을 구하시오.
#print(y_pred.values)
#print('*'*100)
#print(test['Gender'].values)
#error율 구해서 출력하기
err_rate = round(np.mean(np.abs(y_pred.values - test['Gender'].values)), 4)
print(err_rate)
```

##### **02 Used Car Price 데이터 (ols)**
중고차의 가격을 다중회귀분석으로 분석하고자 한다. 'Price'를 종속변수로, 'Year', 'Transmission', 'Mileage', 'Fuel type', 'Tax', 'Mpg', 'Engine Size'를 독립변수로 사용한다. 데이터의 구성은 아래 표와 같다.

`[중고차 가격 데이터 세트 변수 설명]`

|변수        |설명                                        |
|:-----------|:-------------------------------------------|
|year        |생산연도                                    |
|transmission|트랜스미션(Ivtanual, Automatic, Semi-Auto)  |     
|mileage     |주행거리                                    |
|fuel type   |연료 타입(Petr이, Diesel. Hybrid)           |   
|tax         |세금                                        |
|mpg         |연비                                        |
|engine Size |엔진 크기                                   |
|Price       |가격($)                                     |    

(1) 다중회귀분석 모형의 결정계수를 구하시오.    
(2) 독립변수 중 회귀계수가 가장 큰 변수와 그 값을 구하여 튜플 타입으로 출력하시오.    
예시 : ('Weight'. 0.09003569454430438)    
(3) 독립변수 중 p-value가 가장 낮은 변수와 그 값을 구하여 튜플 타입으로 출력하시오.    

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/eyejjang/BigData/main/dataset_yemoon/datasets/Part5/ex7/07.03.02-used_car_price_dataset.csv')
print(df)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['transmission'] = le.fit_transform(df['transmission'])
df['fuelType'] = le.fit_transform(df['fuelType'])
```

```python
#다중회귀분석 모형의 결정계수를 구하시오.
import statsmodels.formula.api as smf
model = smf.ols("price~year+transmission+mileage+fuelType+tax+mpg+engineSize", df).fit()
print(model.rsquared)
print('*'*100)
print(model.summary())
```

```python
#독립변수 중 회귀계수가 가장 큰 변수와 그 값을 구하여 튜플 타입으로 출력하시오. 예시: ('Weight', 0.09003569454430438)
coef = model.params
max_coef = coef.idxmax(), coef.max()
print(max_coef)
print(coef.shape)
```

```python
#독립변수 중 p-value가 가장 낮은 변수와 그 값을 구하여 튜플 타입으로 출력하시오.
p_value = model.pvalues
min_p_value = p_value.idxmin(), p_value.min()
print(min_p_value)
```
