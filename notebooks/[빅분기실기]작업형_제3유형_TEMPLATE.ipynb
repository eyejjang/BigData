```python
#대응 표본 집단의 평균 차이
#H₀: μ ≥ 0 치료 후 혈압(두번쨰 분포)이 치료 전 혈압(첫번째 분포)보다 높거나 같다. (치료 후 혈압과 치료 전 혈압의 차이에 대한 평균은 0보다 크거나 같다)
#H₁: μ < 0 치료 전 혈압(첫번쨰 분포)이 치료 후 혈압(두번쨰 분포)보다 높다. (치료 후에 혈압이 더 낮아졌으므로 치료제의 효과가 있다)
import scipy.stats as stats

ttest1 = stats.ttest_rel(bfr, afr, alternative = 'greater')
#TtestResult(statistic=7.076303701373625, pvalue=0.0010523957839292206, df=4)

ttest2 = stats.ttest_rel(afr, bfr, alternative = 'less')
#TtestResult(statistic=-7.076303701373625, pvalue=0.0010523957839292206, df=4)
```
