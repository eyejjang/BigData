## **(1) Airquality 데이터 (ttest_1samp)**    
주어진 데이터(data/airqualky.csv)에는 1973년 5월에서 9월 사이 153일 동안 뉴욕의 대기질 측정 데이터가 저장되어 있다. 이 기간중풍속(mph)의 평균이 뉴욕의 평균풍속 10 mph와 일치하는지 단일표본 t-검정(One Sample t—Test)을 통해 답하고자 한다. 가설은 아래와 같다.

```
• H0 : μ = 10
• H1 : μ ≠ 10
  - Ozone : 오존 농도  
  - Solar.R : 태양열  
  - Wind : 풍속   
  - Temp : 온도(화씨)   
  - Month : 월    
  - Day : 일
※ 단, 데이터의 Wind 변수는 정규 분포를 만족한다고 가정한다.
```

````python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/eyejjang/BigData/main/dataset_yemoon/datasets/supplement/airquality.csv')
````

````python
print(df.info())
print(df.head())
````

① 표본평균 X bar를 구하시오. (단, 반올림하여 소수점 둘째 자리까지 계산한다.)

````python
result = round(df['Wind'].mean(),2)
print(result)
````

② 위의 가설을 검정하기 위한 검정통계량을 구하시오. (단, 반올림하여 소수점 둘째 자리까지 계산한다.)

````python
import scipy.stats as stats
ttest = stats.ttest_1samp(df['Wind'], 10)
#print(ttest)
stat = round(ttest.statistic,2)
print(stat)
````

③ 위의 통계량에 대한 p-값을 구하고，유의수준 0.05하에서 가설 검정의 결과를 채택/기각 중 하나로 선택 하시오. (단, p-값은 반올림하여 소수점 넷째 자리까지 계산한다.)

````python
pval = round(ttest.pvalue,4)
print(pval)
if ttest.pvalue < 0.05 :
    result = '기각'
else :
    result = '채택'

print(result)

````

## **(2) Chickweight 데이터 (ttest_ind)**    
주어진 데이터(data/chickweight.csv)에는 4가지 종류의 식단으로 구분된 병아리들의 무게를 측정한 데이터가 저장되어 있다. 이 중 1번 식단을 먹은 병아리의 평균 무게가 4번 식단을 먹은 병아리의 평균 무게와 일치 하는지 독립표본 t-검정(Two Sample t-Test)을 통해 답하고자 한다. 가설은 아래와 같다.    
```
• H0 : μ1 = μ4    
• H1 : μ1 ≠ μ4    

- weight : 병아리의 무게     
- Diet : 식단의 종류    

※ 단, 데이터의 weight 변수들은 정규 분포를 만족하며. 두 그룹은 등분산을 띠지 않는다.
```

````python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/eyejjang/BigData/main/dataset_yemoon/datasets/supplement/chickweight.csv')
````

````python
print(df.head())
print(df.info())
print(df.describe(include='all'))
print(df.isnull().sum())
print(df['Diet'].value_counts())
````

① 표본평균 X1 bar, X4 bar를 구하시오. (단，반올림하여 소수점 둘째 자리까지 계산한다.)    

````python
#x1 = df.loc[df['Diet']==1]
#x4 = df.loc[df['Diet']==4]
x1 = df[df['Diet']==1]
x4 = df[df['Diet']==4]
mean_1 = x1['weight'].mean()
mean_4 = x4['weight'].mean()
print(round(mean_1,2))
print(round(mean_4,2))
````

② 위의 가설을 검정하기 위한 검정통계량을 구하시오. (단，반올림하여 소수점 둘째 자리까지 계산한다.)    

````python
import scipy.stats as stats
t_score, p_value = stats.ttest_ind(x1['weight'], x4['weight'], equal_var=False)
print(round(t_score, 2))
````

③ 위의 통계량에 대한 p-값을 구하고，유의수준 0.05하에서 가설 검정의 결과를 채택/기각중 하나로 선택 하시오. (단, p-값은 반올림하여 소수점 넷째 자리까지 계산한다.)    

````python
print(round(p_value, 4), "(", round(p_value, 20), ")")
if p_value < 0.05 :
    result = '기각'
else :
    result = '채택'
print(result)
````

## **(3) Students Performance 1 데이터 (f_oneway)**    
주어진 데이터(data/StudentsPerformancel.csv)에는 학생 1,000명의 인종과 과목별 시험 성적이 담겨져 있다. 인종(A그룹, B그룹, C그룹, D그룹)에 따라 수학 성적에 차이가 있는지 일원분산분석(One-way ANOVA) 통해 답하고자 한다. 가설은 아래와 같다.     

```
• H0 : μA = μB = μC = μD    
• H1 : Not H0    

※ 단, 주어진 데이터가 정규 분포와 등분산을 만족한다고 가정한다.
```

````python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/eyejjang/BigData/main/dataset_yemoon/datasets/supplement/StudentsPerformance1.csv', encoding='cp949')
````

````python
print(df.head())
print(df.info())
print(df.describe(include='all'))
print(df.isnull().sum())
print(df['인종'].value_counts())
````

① 표본평균 XA bar, XB bar, XC bar, XD bar를 구하시오. (단, 반올림하여 소수점 둘째 자리까지 계산한다.)    

````python
df_a = df[df['인종']=='A그룹']
df_b = df[df['인종']=='B그룹']
df_c = df[df['인종']=='C그룹']
df_d = df[df['인종']=='D그룹']
mean_a = df_a['수학 점수'].mean()
print(round(mean_a,2))
mean_b = df_b['수학 점수'].mean()
print(round(mean_b,2))
mean_c = df_c['수학 점수'].mean()
print(round(mean_c,2))
mean_d = df_d['수학 점수'].mean()
print(round(mean_d,2))
````

② 위의 가설을 검정하기 위한 검정통계량을 구하시오. (단, 반올림하여 소수점 둘째 자리까지 계산한다.)    

````python
import scipy.stats as stats
f_score, p_value = stats.f_oneway(df_a['수학 점수'], df_b['수학 점수'], df_c['수학 점수'], df_d['수학 점수'])
print(round(f_score,2))
````

③ 위의 통계량에 대한 p-값을 구하고，유의수준 0.05하에서 가설 검정의 결과를 채택/기각 중 하나로 선택 하시오. (단, p-값은 반올림하여 소수점 넷째 자리까지 계산한다.)    

````python
print(round(p_value, 4))
if p_value < 0.05 :
    result = '기각'
else :
    result = '채택'
print(result)
````

## **(5) Airquality 데이터 (ols)**    

주어진 데이터(data/airquality.csv)에는 1973년 5월에서 9월 사이 153일 동안의 뉴욕의 대기질 측정 데이터가 저장되어 있다. 데이터 중 Solar.R, Wind, Temp, Month가 Ozone에 미치는 영향을 분석하고자 한다. 다중회귀 분석모델을 만들고 다음의 질문에 답하시오. (단, 데이터 전처리 수행 후 진행한다.)

```
- Ozone : 오존 농도     
- Solar.R : 태양열     
- Wind : 풍속     
- Temp : 온도(화씨)     
- Month : 월     
- Day : 일    
```

````python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/eyejjang/BigData/main/dataset_yemoon/datasets/supplement/airquality.csv')
````

````python
print(df.head())
print(df.info())
print(df.describe())
print(df.isnull().sum())
````

① 다중회귀분석 모형의 결정계수를 구하시오.     

````python
dt = df.copy()
#dt = dt.dropna(axis=0)
dt = dt.fillna(dt.median()) # 결측치 최빈값으로 대체
dt.pop('Day')
dt = dt.rename(columns={"Solar.R":"SolarR"})
import statsmodels.formula.api as smf
model = smf.ols('Ozone~SolarR+Wind+Temp+Month', dt).fit()
print(model.rsquared)
print('*'*100)
print(model.summary())
````

② 독립변수 중 회귀 계수가 가장 큰 변수와 값을 구하여 튜플 타입으로 출력하시오.    

````python
coef = model.params
max_coef = coef.idxmax(), coef.max()
print(max_coef)
````

③ 독립변수 중 p-value가 가장 낮은 변수와 값을 구하여 튜플 타입으로 출력하시오.   

````python
pval = model.pvalues
min_pval = pval.idxmin(), pval.min()
#print(pval['Wind'])
#print(pval['Temp'])
print(min_pval)
````

##### **(1) 혈압 치료 전 • 후 데이터 (ttest_rel)**
```
 - 가설  
  H₀: μ ≥ 0 (치료 후 혈압과 치료 전 혈압의 차이에 대한 평균은 0보다 크거나 같다)  
  H₁: μ < 0 (치료 후에 혈압이 더 낮아졌으므로 치료제의 효과가 있다)  
  * μ: (치료 후 혈압 - 치료 전 혈압)의 평균  
```
```
 - 혈압 정보  
  • 치료제 복용 전 5명의 환자들의 혈압 : 200, 210, 190, 180, 175  
  • 치료제 복용 후 5명의 환자들의 혈압 : 180, 175, 160, 150, 160  
```
5명의 환자를 대상으로 치료제를 복용하기 전과 후의 혈압을 측정하였다. 치료제가 효과가 있는지 쌍체표본 T-검정을 통해 답하고자 한다. (단, 표본이 정규성을 만족한다는 가정하여 단측검정 수행)  

````python
bfr = [200, 210, 190, 180, 175]
afr = [180, 175, 160, 150, 160]
print(pd.DataFrame(bfr).mean())
print(pd.DataFrame(afr).mean())
````

**01** 위의 가설을 검정하기 위한 검정통계량을 입력하시오. (반올림하여 소수 다섯째 자리까지 계산)

````python
import scipy.stats as stats
ttest1 = stats.ttest_rel(bfr, afr, alternative = 'greater')
ttest2 = stats.ttest_rel(afr, bfr, alternative = 'less')
#H₀: μ ≥ 0 치료 후 혈압(두번쨰 분포)이 치료 전 혈압(첫번째 분포)보다 높거나 같다. (치료 후 혈압과 치료 전 혈압의 차이에 대한 평균은 0보다 크거나 같다)
#H₁: μ < 0 치료 전 혈압(첫번쨰 분포)이 치료 후 혈압(두번쨰 분포)보다 높다. (치료 후에 혈압이 더 낮아졌으므로 치료제의 효과가 있다)
#two-sided: 샘플의 기초가 되는 분포의 평균이 동일하지 않은 경우
#less: 첫 번째 샘플의 기반이 되는 분포의 평균이 두 번째 샘플의 기반이 되는 분포의 평균보다 작은 경우
#greater: 첫 번째 샘플의 기반이 되는 분포의 평균이 두 번째 샘플의 기반이 되는 분포의 평균보다 큰 경우
stat = round(ttest1.statistic,5)
print(stat)
print('*'*100)
print(ttest1)
print(ttest2)
````

**02** 위의 통계량에 대한 p-value를 구하여 입력하시오. (반올림하여 소수 다섯째 자리까지 계산)

````python
pval = round(ttest.pvalue,5)
print(pval)
````

**03** 유의수준 0.05 하에서 가설검정의 결과를 (채택/기각) 중 하나를 선택하여 입력하시오.

````python
if ttest.pvalue < 0.05 :
    result = '기각'
else :
    result = '채택'
print(result)
````

데이터가 비정규분포인 경우에도 다음의 조건 중 하나에 부합하면 1-표본 t검정을 사용할 수 있다. :

##### **(1) 무작위생성 100개 데이터 (F-test)**
```
 - 가설  
  H₀: 두 그룹의 분산은 같다.
  H₁: 두 그룹의 분산은 같지 않다.  
```
```
 - 데이터
  • A 그룹(df1) : 1, 2, 3, 4, 6
  • B 그룹(df2) : 4, 5, 6, 7, 8
```
무작위로 100개의 값을 생성한 두 그룹(A, B)의 분산이 같은지에 대하여 F-검정을 수행하시오.

**01** 검정 통계량을 소수 3번째 자리에서 반올림하여 소수 2번째 자리까지 출력하시오.

````python
import numpy as np
df10 = np.array([1, 2, 3, 4, 6])
df20 = np.array([4, 5, 6, 7, 8])

df10_var = np.var(df10, ddof=1)
df20_var = np.var(df20, ddof=1)
print(df10_var, df20_var)
import scipy.stats as stats
ftest = stats.f_oneway(df10, df20)
print(ftest)
ttest = stats.ttest_ind(df10, df20, equal_var = False)
print(ttest)


# F 통계량 값을 계산하기 위해 f_test 사용자 정의 함수 생성
def f_test(x, y):
  # F 통계량은 F = S²₁/ S²₂식으로 계산 : 이 때 분자가 분모보다 커야(S²₁> S²₂) 하므로 분모가 더 큰 경우 x, y를 바꿔줌
  if np.var(x, ddof = 1) < np.var(y, ddof = 1):
    x, y = y, x
  # F 통계량 = x의 표본 분산 / y의 표본 분산
  f_value = np.var(x, ddof = 1) / np.var(y, ddof = 1)
  # F 통계량은 표본 분산을 사용하므로 (데이터의 개수 - 1)로 계산해 자유도를 맞춰줌
  x_dof = x.size - 1
  y_dof = y.size - 1
  # F 분포의 누적 분포 함수(f.cdf)를 사용하여 주어진 F-검정 통계량(f_value)과 지정된 자유도(x_dof와 y_dof)를 이용해 p-value 계산
  # 양측 검정을 위해 계산된 값에 2를 곱함
  p_value = round((1 - stats.f.cdf(f_value, x_dof, y_dof)) * 2, 4)
  #p_value = round((1 - stats.f.cdf(f_value, x_dof, y_dof)) + stats.f.sf(f_value, x_dof, y_dof), 4)
  # p-value가 0.05 미만이면 p_result에 "기각"을, 그렇지 않으면 "채택"을 저장
  if p_value < 0.05:
    p_result = "기각"
  else:
    p_result = "채택"
  return f_value, p_value, p_result
# f_test 함수에 df1, df2 값을 전달하면 함수의 반환값 f_value, p_value 값을 받아옴 : 해당 값을 result 변수에 저장
result = f_test(df10, df20)
# 결과 출력
# 통계량 = 1.48, p-value = 0.7133, p-value 0.7133 > 유의수준 0.05 → 귀무가설(두 그룹의 분산은 같음) 채택
print(result)
````

##### **(2) 콘서트장 남 • 녀 데이터 (chisquare)**
```
 - 가설  
  H₀: 콘서트장에서 남성, 여성의 비율은 35%, 65%임
  H₁: 콘서트장에서 남성, 여성의 비율은 35%, 65%가 아님
```
콘서트장에서 남성이 340명, 여성이 540명이 있을 때 남성, 여성 비율이 35%와 65%인지를 카이제곱검정을 이용하여 분석하려고 한다. 적합도 검정을 실시하시오. *이탤릭체 텍스트*

````python
import numpy as pd

# 콘서트장에서 남성 340명, 여성 540명 데이터를 저장한 num 변수 생성
num = np.array([340, 540])

# 기대빈도를 계산(340명 + 540명 중에서 35%가 남성, 65%가 여성)해서 expected 변수에 저장
expected = np.array([0.35, 0.65]) * np.sum(num)

from scipy.stats import chisquare

# 남성:여성 = 35:65 비율에 대한 카이제곱검정을 실행하고 결과를 result 변수에 저장
result = chisquare(f_obs = num, f_exp = expected)

# 카이제곱 검정통계량 = statistic 값 = 5.114885114885115 / 자유도 = 1 / p-value = 0.023721436858355486
print(result)
````
