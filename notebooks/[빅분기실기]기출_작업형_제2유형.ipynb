# **2021년 제2회 기출복원문제**

##### **E-commerce Shipping 데이터 (ROC_AUC)**
다음은 E-commerce Shipping 데이터 세트이다. 주어진 훈련 데이터 세트를 활용하여 고객이 주문한 물품의 정시 도착 여부를 예측하고 해당 예측 결과를 csv 파일로 제출하시오.    
※ 결과 제출 양식 : 제출한 예측값의 ROC_AUC 점수 결과를 통해 영역별 배점에 따라 최종 점수가 반영될 예정    

|Index|y_pred|
|:--:|:--:|
|0|0.2543|
|1|0.1324|
|2|0.5892|
|...|...|

[결과제출양식]

|변수|설명|
|:--:|:--:|
|ID|고객 ID 번호|
|Warehouse_block|창고의 블록 단위 구역(A, B. C. D. F)|
|Mode_of_Shipment|제품 배송 방법|
|Customer_care_calls|문의 전화 수|
|Customer_rating|고객의 등급(1:가장 낮음, 5:가장 높음)|
|Cost_of_the_Product|제품의 비용(달러 기준)|
|Prior_purchases|사전 구매 수량|
|Product_importance|제품의 중요도|
|Gender|성별(F:여성, M:남성)|
|Discount_offered|할인혜택|
|Weight_in_gas|그램 단위 무게|
|Reached.on.Time_Y.N|정시 도착 여부(1:도착하지 않음, 0:정시 도착)|

[E commerce Shipping 데이터 세트 변수 설명]

```python
import pandas as pd
df_train_x = pd.read_csv("https://raw.githubusercontent.com/eyejjang/BigData/main/dataset_yemoon/datasets/Part5/204_x_train.csv")
df_train_y = pd.read_csv("https://raw.githubusercontent.com/eyejjang/BigData/main/dataset_yemoon/datasets/Part5/204_y_train.csv")
df_test_x = pd.read_csv("https://raw.githubusercontent.com/eyejjang/BigData/main/dataset_yemoon/datasets/Part5/204_x_test.csv")
df_test_y = pd.read_csv("https://raw.githubusercontent.com/eyejjang/BigData/main/dataset_yemoon/datasets/Part5/204_y_test.csv")
```

```python
df_train_x.describe()-df_test_x.describe()
```

```python
## 데이터 확인
print(df_train_x.head())
print('*'*100)
print(df_test_x.head())
print('*'*100)
print(df_train_y.head())
print('*'*100)
print(df_test_y.head())
print('*'*100)
```

```python
## 결측치 확인
print(df_train_x.isna().sum())
print('*'*100)
print(df_test_x.isna().sum())
print('*'*100)
print(df_train_y.isna().sum())
print('*'*100)
print(df_test_y.isna().sum())
print('*'*100)
```

```python
print(df_train_x.info())
print('*'*100)
print(df_train_y.info())
```

```python
## 컬럼 그룹화
col_del = ['ID']
col_category = ['Warehouse_block','Mode_of_Shipment','Product_importance','Gender']
col_scale = ['Customer_care_calls','Customer_rating','Cost_of_the_Product','Prior_purchases','Discount_offered','Weight_in_gms']
col_y = ['Reached.on.Time_Y.N']
```

```python
## 라벨인코딩
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for i in col_category :
    le.fit(df_train_x[i])
    df_train_x[i] = le.transform(df_train_x[i])
    df_test_x[i] = le.transform(df_test_x[i])
```

```python
print(df_train_x.head())
print(df_test_x.head())
```

```python
## 스케일링
from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()
mm.fit(df_train_x[col_scale])
df_train_x[col_scale] = mm.transform(df_train_x[col_scale])
df_test_x[col_scale] = mm.transform(df_test_x[col_scale])
```

```python
print(df_train_x.head())
print(df_test_x.head())
```

```python
## 데이터셋 분리
from sklearn.model_selection import train_test_split
#print(help(sklearn.model_selection.train_test_split))
train_x, valdn_x, train_y, valdn_y = train_test_split(df_train_x[col_category+col_scale], df_train_y[col_y], test_size = 0.2, stratify=df_train_y[col_y])
print(train_x.shape)
print(valdn_x.shape)
print(train_y.shape)
print(valdn_y.shape)
print(train_y.value_counts())
print(valdn_y.value_counts())
```

```python
## 랜덤포레스트 모델링
from sklearn.ensemble import RandomForestClassifier
RF_model = RandomForestClassifier()
RF_model.fit(train_x, train_y.values.ravel())
```

```python
## XG부스트 모델링
from xgboost import XGBClassifier
XGB_model = XGBClassifier()
XGB_model.fit(train_x, train_y.values.ravel())
```

```python
## 예측값 확인
y_pred_RF_model_train = RF_model.predict_proba(train_x)[:,1]
y_pred_RF_model_valdn = RF_model.predict_proba(valdn_x)[:,1]
y_pred_XGB_model_train = XGB_model.predict_proba(train_x)[:,1]
y_pred_XGB_model_valdn = XGB_model.predict_proba(valdn_x)[:,1]
```

```python
## ROC_AUC 값 확인
from sklearn.metrics import roc_auc_score
roc_auc_score_RF_train = roc_auc_score(train_y, y_pred_RF_model_train)
roc_auc_score_RF_valdn = roc_auc_score(valdn_y, y_pred_RF_model_valdn)
print('Score_RF_train: ',roc_auc_score_RF_train,'\t Score_RF_valdn: ',roc_auc_score_RF_valdn)
roc_auc_score_XGB_train = roc_auc_score(train_y, y_pred_XGB_model_train)
roc_auc_score_XGB_valdn = roc_auc_score(valdn_y, y_pred_XGB_model_valdn)
print('Score_XGB_train: ',roc_auc_score_XGB_train,'\t Score_XGB_valdn: ',roc_auc_score_XGB_valdn)

## 정시 도착 여부（1 : 도착하지 않음, 0 : 정시 도착）- 문제가 잘못된 것으로 보임. 1 : 정시 도착, 0 : 도착하지 않음 으로 변경이 맞는 것으로 보임
#predict_proba(valdn_x)[:,0]
#Score_RF_train:  0.0 	 Score_RF_valdn:  0.25951538109232475
#Score_XGB_train:  0.006779693872225576 	 Score_XGB_valdn:  0.25702024731880463
#predict_proba(valdn_x)[:,1]
#Score_RF_train:  1.0 	 Score_RF_valdn:  0.7404846189076753
#Score_XGB_train:  0.9932203061277745 	 Score_XGB_valdn:  0.7429797526811954
```

```python
## 결과 데이터셋 생성
y_pred_XGB_model_test = XGB_model.predict_proba(df_test_x[col_category+col_scale])[:,1]
result = pd.DataFrame({'index':df_test_x.index,'y_pred':y_pred_XGB_model_test})
print(result)
```

```python
## 결과파일 저장
result.to_csv('02_02.csv', index=False)
```

```python
## 데이터 확인
print(df_train_x.info())
print(df_train_x.describe(include='all'))
print(df_train_x.head())
print('*'*100)
print(df_train_y.info())
print(df_train_y.describe(include='all'))
print(df_train_y.head())
print('*'*100)
print(df_test_x.info())
print(df_test_x.describe(include='all'))
print(df_test_x.head())
```

```python
## 결측치 확인
print(df_train_x.isna().sum())
print(df_train_y.isna().sum())
print(df_test_x.isna().sum())
```

```python
## 컬럼 그룹화
col_del = ['ID']
col_category = ['Gender','Ever_Married','Graduated','Profession','Spending_Score']
col_scale = ['Age','Work_Experience','Family_Size']
#y = Segmentation
```

```python
## 라벨인코팅
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for i in col_category :
    le.fit(df_train_x[i])
    df_train_x[i] = le.transform(df_train_x[i])
    df_test_x[i] = le.transform(df_test_x[i])

le.fit(df_train_y['Segmentation'])
df_train_y['Segmentation'] = le.transform(df_train_y['Segmentation'])

## 원 핫 인코딩
#df_train_x = pd.get_dummies(df_train_x)
#df_test_x = pd.get_dummies(df_test_x)

print(df_train_x.head())
print(df_test_x.head())
```

```python
## 스케일링
from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()
mm.fit(df_train_x[col_scale])
df_train_x[col_scale] = mm.transform(df_train_x[col_scale])
df_test_x[col_scale] = mm.transform(df_test_x[col_scale])

print(df_train_x.head())
print(df_test_x.head())
```

```python
## 트레이닝셋, 검증셋 분리
from sklearn.model_selection import train_test_split
train_x, valdn_x, train_y, valdn_y = train_test_split(df_train_x[col_category+col_scale], df_train_y['Segmentation'], test_size = 0.2, stratify = df_train_y['Segmentation'])
print(train_x.shape)
print(valdn_x.shape)
print(train_y.shape)
print(valdn_y.shape)
```

```python
## 랜덤포레스트 모델링
from sklearn.ensemble import RandomForestClassifier
RF_model = RandomForestClassifier(random_state=123)
RF_model.fit(train_x, train_y)
```

```python
## XG부스트 모델링
from xgboost import XGBClassifier
XGB_model = XGBClassifier(random_state=123)
XGB_model.fit(train_x, train_y)
```

```python
## 결과값 예측
y_train_RF_model = RF_model.predict(train_x)
y_valdn_RF_model = RF_model.predict(valdn_x)
print(y_train_RF_model)
print(y_valdn_RF_model)
print('*'*100)
y_train_XGB_model = XGB_model.predict(train_x)
y_valdn_XGB_model = XGB_model.predict(valdn_x)
print(y_train_XGB_model)
print(y_valdn_XGB_model)
```

```python
## f1 스코어 확인
from sklearn.metrics import f1_score
print("## F1스코어 확인")
score_RF_train = f1_score(train_y, y_train_RF_model, average='macro')
score_RF_valdn = f1_score(valdn_y, y_valdn_RF_model, average='macro')
print('score_RF_train: ', score_RF_train, '\t score_RF_valdn: ', score_RF_valdn)
score_XGB_train = f1_score(train_y, y_train_XGB_model, average='macro')
score_XGB_valdn = f1_score(valdn_y, y_valdn_XGB_model, average='macro')
print('score_XGB_train: ', score_XGB_train, '\t score_XGB_valdn: ', score_XGB_valdn)
```

```python
##RMSE 확인
from sklearn.metrics import mean_squared_error
import numpy as np
print("## RMSE 확인")
train_model_RF_rmse = np.sqrt(mean_squared_error(train_y, y_train_RF_model))
valdn_model_RF_rmse = np.sqrt(mean_squared_error(valdn_y, y_valdn_RF_model))
print('RMSE_RF_train:', train_model_RF_rmse, '\t RMSE_RF_valdn:', valdn_model_RF_rmse)
train_model_XGB_rmse = np.sqrt(mean_squared_error(train_y, y_train_XGB_model))
valdn_model_XGB_rmse = np.sqrt(mean_squared_error(valdn_y, y_valdn_XGB_model))
print('RMSE_XGB_train:', train_model_XGB_rmse, '\t RMSE_XGB_valdn:', valdn_model_XGB_rmse)
```

```python
y_test_predLR = modelLR.predict(df_test_x[col_category+col_scale])
y_test_predRF = modelRF.predict(df_test_x[col_category+col_scale])
y_test_predXGB = modelXGB.predict(df_test_x[col_category+col_scale])

test_LR_MSE = mean_squared_error(df_test_y['Sales'], y_test_predLR)
test_scoreLR = np.sqrt(test_LR_MSE)
test_RF_MSE = mean_squared_error(df_test_y['Sales'], y_test_predRF)
test_scoreRF = np.sqrt(test_RF_MSE)
test_XGB_MSE = mean_squared_error(df_test_y['Sales'], y_test_predXGB)
test_scoreXGB = np.sqrt(test_XGB_MSE)

print('LogisticRegression - test: ', test_scoreLR)
print('RandomForest - test: ', test_scoreRF)
print('XGBoost - test: ', test_scoreXGB)
```
