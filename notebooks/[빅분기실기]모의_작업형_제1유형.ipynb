## **(1) Iris 데이터**    
다음은 Iris 데이터 세트이다. 아래 문제 01〜05의 답을 구하여라.

|변수|설명|
|:-:|:-:|
|sepal_length|꽃받침 길이|
|sepal_width|꽃받침 너비|
|petal_length|꽃잎 길이|
|petal_width|꽃잎 너비|
|species|꽃종류|

````python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/eyejjang/BigData/main/dataset_yemoon/datasets/Part2/iris.csv')
````

**01** Iris 데이터 세트에서 species 변수 값별로 petal_width 상위 25% 값을 구한 후, 가장 큰 값과 가장 작은 값 사이의 차이를 계산하시오.

````python
df1 = df.groupby(['species']).quantile(0.75)
ans = abs(df1['petal_width'].max() - df1['petal_width'].min())
print(ans)
````

**02** Iris 데이터 세트에서 sepal_length 변수의 전체 값들을 중복 없이 리스트로 만든 후, 오름차순으로 정렬했을 때 리스트의 10번째에 오는 숫자를 계산하시오.

````python
df2 = df['sepal_length'].unique()
ans = pd.DataFrame(data={'sepal_length':df2}).nsmallest(10, 'sepal_length')['sepal_length'].max()
print(ans)
````


**03** Iris 데이터 세트를 sepal_width 변수 값으로 내림차순한 후 위에서부터 100개 행을 학습 데이터로 분리하고, 학습 데이터에서 sepal_width의 표준편차 값을 구한 후 50을 곱한 값을 계산하시오. (단, 최종 계산 값의 소수점 이하는 버린다.)

````python
df3 = df.nlargest(100, 'sepal_width')
ans = int(df3['sepal_width'].std() * 50)
print(ans)
````

**04** Iris 데이터 세트에서 species 변수 값별로 petal_length 상위 10개 행（큰 값）의 평균값을 구한 후, 평균값의 합을 계산하시오. (단, 최종 계산 값의 소수점 이하는 반올림한다.)

````python
#df4 = df.groupby(['species']).nlargest(10, 'petal_length')
df4 = df.sort_values(['species', 'petal_length'], ascending=False)
df5 = df4.groupby(['species']).head(10)
ans = round(df5.groupby(['species'])['petal_length'].mean().sum())
print(ans)
````

**05** Iris 데이터 세트에서 sepal_length 변수가 이상치를 가지는 데이터 행 수를 계산하시오. [단, 이상치 범위는 (평균+2×표준편차) 이상이거나 (평균-2×표준편차)이하이다.]

````python
df6 = df.loc[(df['sepal_length'] > df['sepal_length'].mean() + 2 * df['sepal_length'].std()) | (df['sepal_length'] < df['sepal_length'].mean() - 2 * df['sepal_length'].std())]
ans = len(df6)
print(ans)
````

## **(2) Wine 데이터**    
다음은 Wine 데이터 세트이다. 아래 문제 01〜05의 답을 구하여라.

|변수|설명|
|:-:|:-:|

````python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/eyejjang/BigData/main/dataset_yemoon/datasets/Part2/wine.csv')
````

**01** Wine 데이터 세트에서 total_phenols 결측치를 제거하고 alcalinity_of_ash 변수의 결측치는 중앙값으로 대체한 후, alcalinity_of_ash 평균값을 계산하시오. (단, 최종 계산 값의 소수점 이하는 버린다.)

````python
df1 = df.dropna(subset=['total_phenols'])
df2 = df1.fillna(df1['alcalinity_of_ash'].median())
ans = int(df2['alcalinity_of_ash'].mean())
print(ans)
````

**02** Wne 데이터 세트에서 alcohol 값이 alcohol 평균 값 이상이고, color_intensity 값이 color_intensity 평균 값 이상인 데이터만 남겼을 때, 가장 많은 class 값을 구하시오.

````python
df3 = df.loc[(df['alcohol'] >= df['alcohol'].mean()) & (df['color_intensity'] >= df['color_intensity'].mean())]
df4 = df3.groupby(['class'])['class'].count()
ans = df4.idxmax()
print(ans)
````


**03** Wine 데이터 세트에서 class 1에 해당하는 데이터의 ash 평균값과 Wine 데이터 세트의 모든 결측치를 제거한 후의 ash 값의 평균값의 차이의 절댓값을 계산하시오. (단, 최종 계산 값의 소수점 아래 4번째 자리에서 반올림한다.)

````python
df5 = df.loc[df['class'] == 1]
df6 = df.dropna()
ans = round(abs(df5['ash'].mean() - df6['ash'].mean()), 3)
print(ans)
````

**04** Wine 데이터 세트에서 alcohol 변수의 상위에서 20번째 값 alcohol을 오름차순 정렬했을 때 20번째에 위치한 값으로 상위 20개의 값을 변환한 후, proanthocyanins가 2 이하인 데이터를 추출하여 alcohol의 평균값을 계산하시오.
(단, 최종 계산 값의 소수점 아래 3번째 자리에서 반올림한다.)

````python
df7 = df.sort_values(['alcohol'], ascending=True).reset_index(drop=True)
#df7.loc[0:19, 'alcohol'] = df7.loc[29, 'alcohol']
df7.loc[0:19, 'alcohol'] = df.nsmallest(20, 'alcohol')['alcohol'].max()
ans = round(df7.loc[df7['proanthocyanins'] <= 2, 'alcohol'].mean(), 2)
print(ans)
````

**05** Wine 데이터 세트에서 class 1에 해당하는 데이터의 proline 변수의 제3사분위수와 제1사분위수 차를 구하고, class 2에 해당하는 데이터의 proline 변수의 제3사분위수와 제1사분위수 차를 구하여 두 값의 차이의 절댓값을 계산하시오.

````python
df8 = df.loc[df['class'] == 1, 'proline']
df8q = abs(df8.quantile(0.75) - df8.quantile(0.25))
df9 = df.loc[df['class'] == 2, 'proline']
df9q = abs(df9.quantile(0.75) - df9.quantile(0.25))
ans = abs(df8q - df9q)
print(ans)
````

**01** 데이터에서 'Attrition'은 퇴사여부에 대한 종속변수에 해당한다. 종속변수의 값을 다음과 같이 수치로 변환해 새 컬럼으로 추가하고 각 범주별 레코드 수를 계산하시오. (Yes → 1, No → 0)    

````python
df['Attrition2'] = df['Attrition'].map({"No":0, "Yes": 1})
print(df['Attrition2'].value_counts())
````

**02** 데이터셋의 데이터 타입별 컬럼 개수를 계산하고, 범주형 변수 중 유일한 값을 1개만 가지고 있는 컬럼(변수)를 찾아내어 그 변수를 데이터에서 제거하시오.

````python
df1 = df.copy()
ans = df1.dtypes.value_counts()
print(ans)
df2 = pd.DataFrame(data={'cnt':0}, index=df1.select_dtypes(include='object').columns)
for i in df1.select_dtypes(include='object').columns :
    df2.loc[i, 'cnt'] = len(df1[i].value_counts())
ans = df2.loc[df2['cnt'] == 1]
print(ans)
#df1.pop('Over18')
````

**03** 원래 데이터에서 숫자형인 컬럼(변수)만 추출하여 새로운 데이터프레임을 생성하고, 각 변수간의 상관계수(피어슨 상관계수)를 구하고 상관계수가 0.9 이상인 두 개의 컬럼을 찾아내어 그 변수 중 하나를 제거하시오.

````python
df3 = df.select_dtypes(exclude='object')
df4 = df3.corr()
for i in df4.columns :
    if len(df4.loc[(df4[i] >= 0.9) & (df4[i] < 1.0)][i]) > 0 :
        print(i, '-', df4.loc[(df4[i] >= 0.9) & (df4[i] < 1.0)][i])
#df3.pop('MonthlyIncome')
#df3.pop('JobLevel')
````
