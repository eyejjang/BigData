##### **03 California Housing 데이터**
2번 문항에서 활용한 California Housing 데이터 세트를 그대로 활용한다. 인구 컬럼인 population 항목의 이상값의 합계를 계산하시오. (단, 이상값은 평균에서 1.5x표준편차를 초과하거나 미만인 값의 범위로 정한다.)

````python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/eyejjang/BigData/main/dataset_yemoon/datasets/Part5/203_housing.csv')
````

````python
upper = df['population'].mean() + (1.5 * df['population'].std())
lower = df['population'].mean() - (1.5 * df['population'].std())
df1 = df.loc[(df['population'] > upper) | (df['population'] < lower)]
ans = df1['population'].sum()
print(ans)
````

##### **02 BMI 데이터**
BMI지수는 몸무게(kg)를 키(m)의 제곱으로 나누어 구하며, BMI 값에 따른 비만도 분류는 다음과 같다.

|BMI지수 범위|비만도 분류|
|:--:|:--:|
|18.5 미만|저체중|
|18.5 이상 23 미만|정상|
|23 이상 25 미만|과체중|
|25이상 30 미만|강도비만|
|30 이상|중등도비만|

이때 주어진 bmi 데이터 세트에서 비만도가 정상에 속하는 인원 수와 과체중에 속하는 인원 수의 차이를 정수로 출력 하시오.

````python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/eyejjang/BigData/main/dataset_yemoon/datasets/Part5/502_bmi.csv')
````

````python
df['bmi'] = (df['Weight'] / ((df['Height'] / 100) ** 2))
df1 = df.loc[(df['bmi'] >= 18.5) & (df['bmi'] < 23)]
df2 = df.loc[(df['bmi'] >= 23) & (df['bmi'] < 25)]
ans = int(abs(len(df1) - len(df2)))
print(ans)
````

##### **03 Apartment Prices 데이터**
주어진 데이터 세트는 아파트 가격 데이터다. 아파트 가격을 결정하는 10개의 독립변수와 1개의 종속변수가 있다. 각 독립변수의 이상치를 사분위수를 기준으로 측정하고 이상치가 가장 많은 독립변수의 이상치 개수를 구하시오. 이때 결측치를 먼저 제거한 후 이상치를 구하시오. [단, 이상치는 (1Q-1.5IQR)보다 작거나 (3Q+1.5IQR)보다 큰 값이며, IQR은 Q3에서 Q1을 뺀 값이다.]

```
[X변수]    
• Size : 아파트 크기 (제곱미터)    
• Bedrooms : 침실 수    
• Bathrooms : 욕실 수    
• Floor : 층수     
• Age : 아파트 연식     
• Distance To Center : 도심으로부터의 거리(km)     
• Nearby Schools : 근처 학교 수     
• Nearby Parks : 근처 공원 수      
• Public Transport : 대중교통 접근성      
• Safety Index : 안전 지수

[Y변수]    
• Price : 아파트 가격
```



````python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/eyejjang/BigData/main/dataset_yemoon/datasets/Part5/ex7/07.01.03-apartment_prices_dataset.csv')
````

````python
df1 = df.dropna()
df1.pop('Price')
upper = df1.quantile(0.75) + (1.5 * (df1.quantile(0.75) - df1.quantile(0.25)))
lower = df1.quantile(0.75) - (1.5 * (df1.quantile(0.75) - df1.quantile(0.25)))
df2 = pd.DataFrame(columns=['cnt'], index=df1.columns)
for i in df1.columns :
    cnt = len(df1.loc[(df1[i] > upper[i]) | (df1[i] < lower[i])])
    df2.loc[i] = cnt
ans = df2['cnt'].max()
print(ans)
````
